```python
import datetime as dt
```

```python
import pandas as pd
```

```python
# Get dataframe of boroughs

df = pd.read_csv("taxi_zone_lookup.csv")
df
```

```python
# Create dictionary of boroughs, and build the list of locations for each borough 
# (6 boroughs vs of 265 NY locations)

dfdict = {'EWR': [], 'Queens': [], 'Bronx': [], 'Manhattan': [], 'Staten Island': [], 'Brooklyn': [], 'Unknown': []}

for i in range(len(df)):    
    dfdict[df["Borough"][i]].append(df["LocationID"][i])
    
del df
```

```python
# JANUARY
```

```python
# read yellow_tripdata_2018-01.csv (only pickup and dropoff datetime columns)
# transform pickup and dropoff datetime from string to a datetime columns
date_columns = ['tpep_pickup_datetime', 'tpep_dropoff_datetime']
df1 = pd.read_csv('yellow_tripdata_2018-01.csv', usecols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'PULocationID'], parse_dates = date_columns)
df1
```

```python
type(df1.tpep_pickup_datetime[0]) # verify the transformation for pickup datetime
```

```python
type(df1.tpep_dropoff_datetime[0]) # verify the transformation for dropoff datetime
```

```python
# consider only 2018 for pickup datetime

keep = df1[df1['tpep_pickup_datetime'].dt.year == 2018]
del df1
```

```python
# consider only january for pickup datetime

keep = keep[keep['tpep_pickup_datetime'].dt.month == 1]
```

```python
# same above for dropoff datetime

keep = keep[keep['tpep_dropoff_datetime'].dt.year == 2018]
keep = keep[keep['tpep_dropoff_datetime'].dt.month == 1]
```

```python
# format pickup datetime column, leaving only the hours (selected time slots)

keep['tpep_pickup_datetime'] = keep['tpep_pickup_datetime'].dt.hour
keep
```

```python
# remove dropoff datetime column

keep = keep.drop(labels = "tpep_dropoff_datetime", axis = 1)
keep = keep.groupby(["tpep_pickup_datetime", "PULocationID"]).size()
keep
```

```python
# count total number of trips for each borough, and for each time slot

# this dictionary contains the count for each borough

overall_count = [0]*24

borough_count = {'EWR': [0]*24, 'Queens': [0]*24, 'Bronx': [0]*24, 'Manhattan': [0]*24, 'Staten Island': [0]*24, 'Brooklyn': [0]*24, 'Unknown': [0]*24}

# for every total count in keep_loc, get the corresponding time slot, and LocationID
# check what borough the location belongs to, and 
# add the count for that location to the correct borough and time slot

for i in range(0, len(keep)):
    time_slot = keep.index[i][0]
    LocationID = keep.index[i][1]
    for borough in dfdict:
        if LocationID in dfdict[borough]:
            borough_count[borough][int(time_slot)] = borough_count[borough][int(time_slot)] + int(keep[time_slot][LocationID])

    overall_count[int(time_slot)] = overall_count[int(time_slot)] + int(keep[time_slot][LocationID])
```

```python
df_time_slot = pd.DataFrame(list(range(24)), columns = ["Time slots"])
df_overallcount = pd.DataFrame(overall_count, columns = ["NY: N° of trips"])
merge = df_time_slot.join(df_overallcount)
merge
```

```python
# Plot an histogram

barplot = merge.plot(x="Time slots", y="NY: N° of trips", kind='bar', figsize = (15, 6), fontsize = 12)
```

```python
df_count = pd.DataFrame(borough_count["EWR"], columns = ["EWR: N° of trips"])
merge = df_time_slot.join(df_count)
merge
```

```python
# Plot an histogram

barplot = merge.plot(x="Time slots", y="EWR: N° of trips", kind='bar', figsize = (15, 6), fontsize = 12)
```

```python
df_time_slot = pd.DataFrame(list(range(24)), columns = ["Time slots"])
df_count = pd.DataFrame(borough_count["Queens"], columns = ["Queens: N° of trips"])
merge = df_time_slot.join(df_count)
merge
```

```python
# Plot an histogram

barplot = merge.plot(x="Time slots", y="Queens: N° of trips", kind='bar', figsize = (15, 6), fontsize = 12)
```
